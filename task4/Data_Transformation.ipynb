```python
import pandas as pd
df = pd.DataFrame({"A": [12, 4, 5, None, 1],
                  "B": [7, 2, 54, 3, None],
                  "C": [20, 16, 11, 3, 8],
                  "D": [14, 3, None, 2, 6]})

index_ = ['Row_1', 'Row_2', 'Row_3', 'Row_4', 'Row_5']

df.index = index_
df
```

```python
result = df.transform(func =lambda x : x+ 10)
result
```

```python
text = 'thiS iS aN pYTHON clASS'

print("\nConvert String to Upper case:")
print(text.upper())

print("\nConvert String to Lower case:")
print(text.lower())

print("\nConvert first letter to Upper in every word:")
print(text.title())

print("\nString case Swapping:")
print(text.swapcase())

print("\nConvert String first letter to uper:")
print(text.capitalize())

print("\nOriginal String:")
print(text)
```
